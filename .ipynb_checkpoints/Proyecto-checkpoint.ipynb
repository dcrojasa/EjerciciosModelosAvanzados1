```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
##
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
```

```python
df = pd.read_csv('../EjerciciosModelosAvanzados1/BaseTransporte 1.csv', sep=';')
df.head()
```

```python
# seleccionar y convertir a fecha el indice
df1 = df[['Fecha','Sillas']]
df1.set_index('Fecha', inplace=True)
df1.head()
```

```python
# Grafico de los datos
df1.plot(figsize=(50,15), linewidth=1, fontsize=40, color=('b'))
plt.xlabel('Tiempo', fontsize=50)
plt.suptitle('Serie de datos - Sillas', fontsize=40)
plt.legend(prop={'size':40});
```

```python
# Anaisis de tendencia
df1[['Sillas']].rolling(200).mean().plot(figsize=(50,15), linewidth=1, fontsize=40, color=('b'))
plt.xlabel('Tiempo', fontsize=50)
plt.suptitle('Tendencia de datos', fontsize=40)
plt.legend(prop={'size':40});
```

```python
# Analisis estacionalidad
df1.diff().plot(figsize=(50,15), linewidth=1, fontsize=40, color=('b'))
plt.xlabel('Tiempo', fontsize=50)
plt.suptitle('Estacionalidad', fontsize=40)
plt.legend(prop={'size':40});
```
