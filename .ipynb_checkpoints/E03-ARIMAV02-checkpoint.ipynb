# 03.1 - Plot the ACF and PACF of the 4 series in data_arma.csv
### Try to say to which AR or MA process each series belong

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf

df = pd.read_csv('../EjerciciosModelosAvanzados1/data_arma.csv', index_col=0)
df.head()
```

```python
df.info()
```

## 1. Análisis  ts1


La autocorrelación y la autocorrelación parcial muestra que los datos se comprtan cómo un MA de orden 1 (Moving average) ya que  después del lag 1 bajan fuertemente los valores de autocorrelación


```python
df['ts1'].plot()
plot_acf(df['ts1'], lags=50)
plot_pacf(df['ts1'], lags=50);
```

### 2. Análisis ts2
Gráfico de lso datos: No se identifica claramente tendencia, parece no tener

respecto a la autocorrelación y la autocorrelación parcial muestra que los datos se comprtan cómo un AR de orden 1, esto por que la primera gráfica nos muestra que la disminución en la autocorrelación es gradual, por otro lado la autocorrelación parcial nos confirma que es de orden 1. 


```python
df['ts2'].plot()
plot_acf(df['ts2'], lags=50);
plot_pacf(df['ts2'], lags=50);
```

### 3. Análisis ts3
Gráfico de lso datos: No se observa uan tendencia clara, también parece tener una autocorrelación de un orden mayor a 1

Observando los g´raficos de autocorrelación se puede ver que claramente existe una autocorelación y que se comporta como un AR(7), el orden lo obtenemos del gráfico de autocorrelación parcial. 


```python
df['ts3'].plot()
plot_acf(df['ts3'], lags=50);
plot_pacf(df['ts3'], lags=50);
```

### 4. Análisis ts4

# 03.2 - Using the shampoo dataset:
## Find the parameters p, d q that minimizes the mse of the rolling forecast


```python
import pandas as pd
import matplotlib.pyplot as plt

from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
def parser(x):
    return pd.datetime.strptime('190'+x, '%Y-%m')
 
series = pd.read_csv('../EjerciciosModelosAvanzados1/shampoo.csv', header=0, parse_dates=[0], index_col=0, squeeze=True, date_parser=parser)
print(series.head())
series.plot(figsize=(10, 10));
```

```python
X = series.values
size = int(len(X) * 0.66)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = list()
```

```python
# fit model
    # No todas las combinaciones de los parámetros son factibles.
for ma in range(7):
    for ar in range(ma + 1, 8):
        # Predicciones y sobre test
        X = series.values
        size = int(len(X) * 0.66)
        train, test = X[0:size], X[size:len(X)]
        history = [x for x in train]
        predictions = list()
        for t in range(len(test)):
            model = ARIMA(history, order=(ar, 1, ma))
            model_fit = model.fit(disp=0)
            output = model_fit.forecast()
            yhat = output[0]
            predictions.append(yhat)
            obs = test[t]
            history.append(obs)
        error = mean_squared_error(test, predictions)     
        print('p:', ar, '- d: 1 - q:', ma,'- Test MSE: %.3f' % error) 
```
